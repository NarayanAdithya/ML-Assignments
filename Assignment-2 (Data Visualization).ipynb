```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#Loading Dataset
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#Seaborn Pairplot
sns.pairplot(iris,hue='species')
```

Create a kde plot of sepal_length versus sepal width for setosa species of flower.

```python
#Data showing species setosa
new=iris[iris['species']=='setosa']
#Seaborn KDE Plot
sns.kdeplot(new['sepal_width'],new['sepal_length'],shade=True,cmap='plasma',shade_lowest=False)
```

```python
#Using tips dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Using Seaborn FacetGrid
f=sns.FacetGrid(data=tips,row='sex',col='time')
#Mapping grpahs to the FacetGrid graphs
f=f.map(plt.hist,'total_bill',bins=15,color='blue')
```

```python
#Using Seaborn BoxPlot
plt.figure(figsize=(8,8))
sns.boxplot(x=tips['day'],y=tips['total_bill'],hue=tips['sex'])
```

```python
#Using Seaborns Jointplot
sns.set_style('whitegrid')
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex')
```

```python
#Using Seaborns jointplot
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
#The x and y values for the above plot
x=np.array([1,1.5,2,2.5,3,3.5,4])
y=np.array([10,15,20,25,30,32.5,35])
fig=plt.figure(figsize=(6,5))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
#plotting additional 3 points
ax.plot(1.3,9,marker='^')
ax.plot(3.8,25,marker='^')
ax.plot(2.5,27,marker='^')
```

```python
#To replicate the above sin,cos graph
x=np.arange(0,10,0.1)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,label='Sine Plot')
axes.plot(x,z,linestyle='--',label='Cosine Plot')
fig.legend()
```

```python
#The above graphs are subplots, 1st subplot is the e^x and y=0, the second subplot has e^x and y=x plotted in logarithmic y scale
#Declaring x,y,z
x=np.arange(0.1,10,0.1)
y=np.exp(x)
z=np.zeros(len(x))

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
axes[0].plot(x,y)
axes[0].plot(x,z)
axes[1].set_yscale('log')
axes[1].plot(x,y)
axes[1].plot(x,x)
axes[0].set_title('Normal Scale')
axes[1].set_title('Logarithmic Scale')

```

```python
#The above grpah is a y=x^3 graph,the axis have been centered
#Plotting the graph
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=np.arange(-1,1.1,0.01)
y=x**3
ax.plot(x,y)
#Adjusting the Position of the Axes
ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
```
